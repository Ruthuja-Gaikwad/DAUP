```python
import kagglehub
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
file_path = "/root/.cache/kagglehub/datasets/awaiskaggler/insurance-csv/versions/1/insurance.csv"
data = pd.read_csv(file_path)
```

```python
print(data.head())
```

```python
age_vector = data['age']
```

```python
smoker_factor = pd.Categorical(data['smoker'])
```

```python
sample_list = [age_vector[:5].tolist(), smoker_factor[:5].tolist()]
```

```python
matrix = np.array(data[['age', 'bmi', 'expenses']])
```

```python
print("Sample List:")
print(sample_list)
print("\nMatrix:")
print(matrix)
print("age_vector:")
print(age_vector)
print("smoker_factor:")
print(smoker_factor)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['age'], data['expenses'], c='blue', alpha=0.6, label='All Data')
plt.title('Scatter Plot of Expenses vs Age')
plt.xlabel('Age')
plt.ylabel('Expenses')
plt.legend()
plt.grid(True)
plt.show()
```

```python
filtered_data = data[data['smoker'] == 'yes']
```

```python

plt.figure(figsize=(10, 6))
plt.scatter(filtered_data['age'], filtered_data['expenses'], c='red', alpha=0.6, label='Smokers')
plt.title('Scatter Plot of Expenses vs Age (Filtered for Smokers)')
plt.xlabel('Age')
plt.ylabel('Expenses')
plt.legend()
plt.grid(True)

plt.xlim(filtered_data['age'].min() - 5, filtered_data['age'].max() + 5)
plt.ylim(filtered_data['expenses'].min() - 1000, filtered_data['expenses'].max() + 1000)

max_charge = filtered_data.loc[filtered_data['expenses'].idxmax()]
plt.text(max_charge['age'], max_charge['expenses'], 'Highest Charge', fontsize=9, color='green')

plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['age'], data['smoker'], c='blue', alpha=0.6, label='All Data')
plt.title('Scatter Plot of Smoke vs Age')
plt.xlabel('Age')
plt.ylabel('Smoker')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(15, 6))
plt.scatter(data['age'], data['bmi'], c='blue', alpha=0.6, label='All Data')
plt.title('Scatter Plot of Smoke vs Age')
plt.xlabel('Age')
plt.ylabel('BMI')
plt.legend()
plt.grid(True)
plt.show()
```

```python
a = data[data['age'] > 39]['smoker'].value_counts()
b = data[data['age'] <= 39]['smoker'].value_counts()

plt.figure(figsize=(8, 6))

plt.bar(['Above 39', 'Below 39'], [a.sum(), b.sum()], color=['skyblue', 'lightcoral'])

plt.xlabel("Age Group")
plt.ylabel("Number of Smokers")
plt.title("Number of Smokers Above and Below 39")

plt.show()
```
